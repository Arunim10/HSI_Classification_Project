```python
train = {
    'ID':[],
    'Experiment':[],
    'Mineral':[],
    'HSI_Path':[],
    'MASK_Path':[],
    'Spatial_Split':[]
}
test = {
    'ID':[],
    'Experiment':[],
    'Mineral':[],
    'HSI_Path':[],
    'MASK_Path':[],
    'Spatial_Split':[]
}
minerals = df['Mineral'].unique()
for mineral in tqdm(minerals,total=len(minerals)):
    mineral_df = df[df['Mineral']==mineral]
    if mineral_df.shape[0]>1:
        
        ## Train
        train['ID'].append(mineral_df.iloc[0]['ID'])
        train['Mineral'].append(mineral_df.iloc[0]['Mineral'])
        train['Experiment'].append(mineral_df.iloc[0]['Experiment'])
        train['HSI_Path'].append(mineral_df.iloc[0]['HSI_Path'])
        train['MASK_Path'].append(mineral_df.iloc[0]['MASK_Path'])
        train['Spatial_Split'].append(0)
        
        ## Test
        for i,row in mineral_df.iterrows():
            if i==0:
                continue
            test['ID'].append(row['ID'])
            test['Mineral'].append(row['Mineral'])
            test['Experiment'].append(row['Experiment'])
            test['HSI_Path'].append(row['HSI_Path'])
            test['MASK_Path'].append(row['MASK_Path'])
            test['Spatial_Split'].append(0)
    else:
        ## Train
        train['ID'].append(mineral_df.iloc[0]['ID'])
        train['Mineral'].append(mineral_df.iloc[0]['Mineral'])
        train['Experiment'].append(mineral_df.iloc[0]['Experiment'])
        train['HSI_Path'].append(mineral_df.iloc[0]['HSI_Path'])
        train['MASK_Path'].append(mineral_df.iloc[0]['MASK_Path'])
        train['Spatial_Split'].append(1)
        
train_df = pd.DataFrame(train)
test_df = pd.DataFrame(test)
```

```python
from scipy.io import loadmat
import numpy as np
train_test_dict = {}
for _,row in tqdm(train_df.iterrows(),total=len(train_df)):
    ID = row['ID']
    EXP = row['Experiment']
    key = f"{ID}_{EXP}"
    hsi_path = row['HSI_Path']
    hsi_shape = np.transpose(loadmat(hsi_path)['HDR'],axes=(1,2,0)).shape
    spatial_split = row['Spatial_Split']
    train_test_dict[key] = ["Train",hsi_shape,spatial_split]
    
for _,row in tqdm(test_df.iterrows(),total=len(test_df)):
    ID = row['ID']
    EXP = row['Experiment']
    key = f"{ID}_{EXP}"
    hsi_path = row['HSI_Path']
    hsi_shape = np.transpose(loadmat(hsi_path)['HDR'],axes=(1,2,0)).shape
    spatial_split = row['Spatial_Split']
    train_test_dict[key] = ["Test",hsi_shape,spatial_split]
```

```python
train_test_dict_broadcast = spark.sparkContext.broadcast(train_test_dict)
@pandas_udf("string")
def train_test_split_udf(ID,EXP,patch_id):
    # Access the broadcast dictionary
    train_test_dict_broadcast = train_test_dict_broadcast.value
    
    def determine_type(ID, EXP, patch_id):
        data_type = ""
        spatial_split = train_test_dict[f"{ID}_{EXP}"][2]
        
        if spatial_split == 0:
            data_type = train_test_dict[f"{ID}_{EXP}"][0]
        else:
            shape = train_test_dict[f"{ID}_{EXP}"][1]
            x_hsi, y_hsi = patch_id // shape[1], patch_id % shape[1]
            middle_x_hsi = shape[0] // 2 if shape[0] % 2 != 0 else shape[0] // 2 - 1
            
            if x_hsi <= middle_x_hsi:
                data_type = "Train"
            else:
                data_type = "Test"
        
        return data_type
    
    # Apply function to each row
    result = [determine_type(ID[i], EXP[i], patch_id[i]) for i in range(len(ID))]
    return pd.Series(result)
```
