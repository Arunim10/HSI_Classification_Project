```python
from glob import glob
import os
import h5py
from tqdm import tqdm

hdr_files = sorted(list(glob("D:\\Machine Learning\\Texmin_HSI\\HSI_Classification_Updated\\Minerals_Dataset\\HDR_Data\\Extracted_Folder\\*\\*.hdr.h5")))
masked_hdr_files = sorted(list(glob("D:\\Machine Learning\\Texmin_HSI\\HSI_Classification_Updated\\Minerals_Dataset\\Masked_HDR_Data\\Extracted_Folder\\*\\*.mhdr.h5")))

for hdr_path,mask_path in zip(hdr_files,masked_hdr_files):
    print(hdr_path.split("\\")[-2:]," Mask ",mask_path.split("\\")[-2:])

```

```python
# Both should be equal to 152 if background image is not downloaded
print(len(hdr_files))
print(len(masked_hdr_files))
```

```python
hdr_shape = []
for hdr_path in tqdm(hdr_files,total=len(hdr_files)):
    data = h5py.File(hdr_path,'r')['hdr']
    hdr_shape.append(data.shape)

mhdr_shapes = []
for mask_path in tqdm(masked_hdr_files,total=len(masked_hdr_files)):
    data = h5py.File(mask_path,'r')['hdr']
    mhdr_shapes.append(data.shape)
    
print(hdr_shape,"\n")
print(mhdr_shapes)
```

```python
import matplotlib.pyplot as plt
hdr = h5py.File(hdr_files[0],'r')['hdr']
mask = h5py.File(masked_hdr_files[0],'r')['hdr']
fig,ax = plt.subplots(1,2,figsize=(10,10))
ax[0].imshow(hdr[0,:,:])
ax[1].imshow(mask[0,:,:])
```
